```python
county_data = pd.read_csv('county_data_2000-2018_geolocate.csv')
county_data["display"] = county_data["county"] + ", " + county_data["state"]
```

```python
fig = px.density_mapbox(county_data, lat='lat', lon='lon', z='filings', radius=30,)
fig.update_layout(mapbox_style="open-street-map")
```
